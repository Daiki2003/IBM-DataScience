```python
books = []
borrowed = []
members = []

def add_books(name):
    books.append(name)
    
def borrow_book(name):
    if name in books:
        books.remove(name)
        borrowed.append(name)
    else:
        print("\nSorry but this book is not available in the moment")
        
def add_member(name):
    members.append(name)
    
def Report():
    print("\nThe available books are: \n")
    for i in books:
        print(i + "\n")
    print("\nThe borrowed books are: \n")
    for i in borrowed:
        print(i + "\n")
    print("\nThe Members in this library are: \n")
    for i in members:
        print(i + "\n")
        
def Return(name):
    if name in borrowed:
        borrowed.remove(name)
        books.append(name)
    else:
        print("\nSorry but this book is not borrowed from us")
```

```python
while True:
    
    operation = input("\nWhat Do you want to do? \n")
    
    if operation == "Add book":
        book = input("\nPlease enter the name of the book \n")
        add_books(book)
        
    elif operation == "Borrow":
        book = input("\nPlease enter the name of the book \n")
        borrow_book(book)
        
    elif operation == "Add member":
        member = input("\nPlease enter your name \n")
        add_member(member)
        
    elif operation == "Report":
        Report()
        
    elif operation == "Return":
        book = input("\nPlease enter the name of the book \n")
        Return(book)
        
    elif operation == "Stop":
        break
        
    elif operation == "Help":
        print("\nThe available functions are : Add Book, Borrow, Add member, Report, Return, Stop")
        
    else:
        print("\nPlease enter a valid function \nType Help for more info")
```
